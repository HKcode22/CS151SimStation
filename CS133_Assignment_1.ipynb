```python
#Load a dataset from GitHub and perform some preliminary evaluation on it.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings("ignore")


# Load the dataset from GitHub
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv'
df = pd.read_csv(url)

# display  first few rows of the dataset
print(df.head())

# get information about the dataset
print(df.info())

# get summary statistics of the dataset
print(df.describe())

# check for missing values
print(df.isnull().sum())

# check for duplicate rows
print(df.duplicated().sum())

# check for unique values in each column
print(df.nunique())

# # check for correlation between columns
# print(df.corr())

# check for outliers in numerical columns
print(df.boxplot())

# check for class imbalance in categorical columns
print(df['sex'].value_counts())
print(df['smoker'].value_counts())
print(df['day'].value_counts())
print(df['time'].value_counts())

# check for missing values
print(df.isnull().sum())

# check for duplicate rows
print(df.duplicated().sum())

# check for unique values in each column
print(df.nunique())

# # check for correlation between columns
# print(df.corr())

# check for outliers in numerical columns
print(df.boxplot())

# check for class imbalance in categorical columns
print(df['sex'].value_counts())
print(df['smoker'].value_counts())
print(df['day'].value_counts())
print(df['time'].value_counts())



#plotting histogram for numerical columns
df.hist(bins=20,figsize=(10,6))
plt.show()

#scatter plot of total bill vs tip
df.plot(kind='scatter', x='total_bill',y='tip', alpha=0.5)
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()

#box plot of total bill by day
df.boxplot(column='total_bill', by='day')
plt.title('Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()

#bar plot of total bill by time
df.groupby('time')['total_bill'].mean().plot(kind='bar')
plt.title('Total Bill by Time')
plt.xlabel('Time')
plt.ylabel('Total Bill')
plt.show()

#pie chart of sex distribution
df['sex'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Sex Distribution')
plt.ylabel('')
plt.show()

#box plot of tips by day of the week
sns.boxplot(x='day',y='tip', data=df)
plt.title('Tips by Day of the Week')
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```
